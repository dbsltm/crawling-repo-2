```python
%%writefile chatbot_insta/chatbot_insta.py
from flask import *

import libs.slack as slack

app = Flask(__name__)


webhook_url = "url 입력"

@app.route("/")
def index():
    return "server is running!!!"

@app.route("/map")
def map():
    return render_template('daum_map_origin.html')

@app.route("/3flobby")
def threef():
    return render_template('3f_html')

@app.route("/bot", methods=["POST"])
def bot():
    
    username = request.form.get("user_name")
    token = request.form.get("token")
    text =  request.form.get("text")
    
    print(username, token, text)
    
    text = text.replace("[bot] ", "")
    
    if text.find(":") < 0:
        slack.send_msg(webhook_url, 
                       """
                       (지역구):(뷰맛집 or 메뉴맛집)과 같이 입력해주세요.
                       ex)[bot] 강남:메뉴맛집
                       
                       """)
        return Response(), 200
    
  
    comm, data = text.split(":")[0], text.split(":")[1]

   
    if comm == "강동":
        
        if data =='메뉴맛집':
            msg = '페더커피☕'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)
       
    elif comm == "강남":
        
        if data =='뷰맛집':
            msg = '카멜커피'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '☕ 맘마미아, 청수당갤러리, 룸서비스301, 노티드청담, 겟썸커피'
            slack.send_msg(webhook_url, msg)
    
    elif comm == "강북":
        
        if data =='뷰맛집':
            msg = '찾지 못했습니다 😢, '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '☕ 카페도모'
            slack.send_msg(webhook_url, msg)
    
    elif comm == "강서":
        
        if data =='메뉴맛집':
            msg = '찾지 못했습니다 😢, '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '헤리스헤이스'
            slack.send_msg(webhook_url, msg)   
            
    elif comm == "관악":
        
        if data =='메뉴맛집':
            msg = '☕포말커피'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)             

    elif comm == "광진":
        
        if data =='메뉴맛집':
            msg = '☕벨롱에스프레소, '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)
            
    elif comm == "구로":
        
        if data =='메뉴맛집':
            msg = '☕이미커피로스터스, '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)   

    elif comm == "금천":
        
        if data =='메뉴맛집':
            msg = '☕카페스미다 '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)  
            
    elif comm == "노원":
        
        if data =='메뉴맛집':
            msg = '☕던모스'
            slack.send_msg(webhook_url, msg)
            
    
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)   
    
    
    elif comm == "동대문":
        
        if data =='뷰맛집':
            msg = '망우로30'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)   
    
    elif comm == "동작":
        
        if data =='메뉴맛집':
            msg = '☕터방내, 시너리, 펠어커피'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)
    
    elif comm == "마포":
        
        if data =='뷰맛집':
            msg = '서울콜렉터'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '☕ 카페평상시, 사월의 물고기, 비포그레이'
            slack.send_msg(webhook_url, msg)       
    
    elif comm == "서대문":
        
        if data =='뷰맛집':
            msg = '오디너리핏'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '☕ 푸어링아웃'
            slack.send_msg(webhook_url, msg)
   
    elif comm == "서초":
        
        if data =='메뉴맛집':
            msg = '☕ 꼼레떼, 헬무트'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)  
    
    elif comm == "성동":
        
        if data =='메뉴맛집':
            msg = '☕ 마를리, 옹근달, 플디,자그마치, 할아버지공장, 소소하게 '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '맛차차, 텅플래닛, 대림창고'
            slack.send_msg(webhook_url, msg)  

    elif comm == "성북":
        
        if data =='메뉴맛집':
            msg = '☕ 마이버터드림 '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '찾지 못했습니다 😢'
            slack.send_msg(webhook_url, msg)

    elif comm == "송파":
        
        if data =='메뉴맛집':
            msg = '☕ 오린지, 사브레, 코히루, 프로퍼커피바 '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '뷰클랜드, 만옥당, '
            slack.send_msg(webhook_url, msg)

    elif comm == "양천구":
        
        if data =='메뉴맛집':
            msg = '찾지 못했습니다 😢 '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '에프스토리 '
            slack.send_msg(webhook_url, msg)            

    elif comm == "용산":
        
        if data =='메뉴맛집':
            msg = '☕ 라라브레드,  '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '김메주와고양이들, 3Flobby, mtl한남, 사유, 엔트러사이트, 음레코드, 챔프커피, 데이로우 '
            slack.send_msg(webhook_url, msg)
            
    elif comm == "종로":
        
        if data =='메뉴맛집':
            msg = '☕ 리틀버틀러, 아뻬서울, 끼룩하우스, 서울커피익선점, 하이웨스트'
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '대충유원지, 서울상회, 호아드'
            slack.send_msg(webhook_url, msg)  

    elif comm == "중구":
        
        if data =='메뉴맛집':
            msg = '찾지 못했습니다 😢 '
            slack.send_msg(webhook_url, msg)
            
        else:
            msg= '카페적당, 더피터커피, 커피사마리아, 클래직'
            slack.send_msg(webhook_url, msg) 
            
    elif comm == "중랑" or "은평" or "영등포" or "도봉":
        
        msg = '지역구에 있는 카페를 찾지 못했습니다 😢 '
        slack.send_msg(webhook_url, msg)            
    
    
    else:
        msg = "{}은(는)없는 명령입니다.".format(comm)
        slack.send_msg(webhook_url, msg)     

    
    
    return Response(), 200


app.run(debug=True)
```
